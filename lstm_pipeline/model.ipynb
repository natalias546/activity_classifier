```python
watch = spark.read.parquet("data/watch_features.parquet")
phone = spark.read.parquet("data/phone_features.parquet")


# get unique seq_ids from watch (same subjects exist in phone)
all_ids = watch.select("seq_id").distinct()

# split seq_ids 60/20/20
train_ids, val_ids, test_ids = all_ids.randomSplit([0.6, 0.2, 0.2], seed=42)

# apply same splits to both sources
watch_train = watch.join(train_ids, on="seq_id")
watch_val   = watch.join(val_ids,   on="seq_id")
watch_test  = watch.join(test_ids,  on="seq_id")

phone_train = phone.join(train_ids, on="seq_id")
phone_val   = phone.join(val_ids,   on="seq_id")
phone_test  = phone.join(test_ids,  on="seq_id")

# sanity check
print("watch train:", watch_train.count())
print("watch val:  ", watch_val.count())
print("watch test: ", watch_test.count())
print("phone train:", phone_train.count())
print("phone val:  ", phone_val.count())
print("phone test: ", phone_test.count())
```
